```python
# Импорт библиотеки
import numpy as np
```

```python
# 1. Создание массивов
print("1. СОЗДАНИЕ МАССИВОВ")
# Из списка
arr1 = np.array([1, 2, 3, 4, 5])
print(f"arr1 из списка: {arr1}")
```

```python
# Массив нулей
zeros_arr = np.zeros(5)
print(f"Массив нулей: {zeros_arr}")

# Массив единиц
ones_arr = np.ones((3, 3))  # 2D массив 3x3
print(f"Массив единиц 3x3:\n{ones_arr}")
```

```python
# Массив с помощью arange
range_arr = np.arange(0, 10, 2)  # от 0 до 10 с шагом 2
print(f"arange от 0 до 10 с шагом 2: {range_arr}")
```

```python
print("\n2. АТРИБУТЫ МАССИВА")
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(f"Матрица:\n{matrix}")
print(f"Форма (shape): {matrix.shape}")
print(f"Тип данных (dtype): {matrix.dtype}")
print(f"Размерность (ndim): {matrix.ndim}")
```

```python
# 3. Индексирование и срезы
print("\n3. ИНДЕКСИРОВАНИЕ И СРЕЗЫ")
arr = np.array([10, 20, 30, 40, 50, 60])
print(f"Исходный массив: {arr}")
print(f"Первый элемент: {arr[0]}")
print(f"Последний элемент: {arr[-1]}")
print(f"Срез с 2 по 4 элемент: {arr[2:5]}")
print(f"Каждый второй элемент: {arr[::2]}")
```

```python
# 4. Векторные операции
print("\n4. ВЕКТОРНЫЕ ОПЕРАЦИИ")
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(f"Массив a: {a}")
print(f"Массив b: {b}")
print(f"Поэлементное сложение: {a + b}")
print(f"Поэлементное умножение: {a * b}")
print(f"Скалярное умножение: {a * 10}")
```

```python
# 5. Статистики
print("\n5. СТАТИСТИКА")
data = np.array([2.5, 3.0, 4.1, 5.2, 1.8])
print(f"Данные: {data}")
print(f"Среднее значение: {np.mean(data):.2f}")
print(f"Стандартное отклонение: {np.std(data):.2f}")
print(f"Минимальное значение: {np.min(data)}")
print(f"Максимальное значение: {np.max(data)}")
print(f"Сумма всех элементов: {np.sum(data)}")
```
